# Plotting x and y points


The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.

# Add Grid Lines to a Plot

```python
xpoints = np.array([10, 20, 35, 40, 50])
ypoints = np.array([60, 55, 45, 70, 75])

plt.figure(figsize=(4,4))

plt.xlabel("x values")
plt.ylabel("Y values")

plt.title("This is plot for adding GRID to plot")
plt.plot(xpoints, ypoints)

plt.grid()

plt.show()
```

# Matplotlib Subplot

```python
# With the subplot() function you can draw multiple plots in one figure:
```

```python
import matplotlib.pyplot as plt
import numpy as np


# first plot

x = np.array([0, 1, 2, 3])
y = np.array([2, 3, 5, 8])


plt.subplot(2, 3, 1)

plt.plot(x, y)
plt.plot(x, y)

# second plot

x = np.array([2, 4, 5, 8])
y = np.array([7, 10, 15, 20])

plt.subplot(2, 3, 2) #plt.subplot(2, 3, 1) --> total number of rows, total number columns & plot number

plt.plot(x, y)

# third plot
x = np.array([0, 1, 2, 4])
y = np.array([3, 8, 2, 10])

plt.subplot(2, 3, 3)

# forth plot

x = np.array([1, 7, 3, 6])
y = np.array([3, 4, 8, 10])

plt.subplot(2, 3 ,4)
plt.plot(x, y)


# fifth plot

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 2, 10])

plt.subplot(2, 3, 5)
plt.plot()

# sixth plot

x = np.array([1, 1, 4, 3])
y = np.array([10, 9, 30, 40])

plt.subplot(2, 3, 6)
plt.plot()

# ax.remove()
plt.show()
```

# Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([4, 10, 8, 7])

plt.figure(figsize=(3, 3))
plt.title("Bar Graph")

plt.bar(x, y, color = "pink")

plt.show()
```

# Matplotlib Bars

Horizontal Bars - 
If you want the bars to be displayed horizontally instead of vertically, use the barh() function:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([4, 10, 8, 7])

plt.figure(figsize=(3, 3))
plt.title("Bar Graph")

plt.barh(x, y, color = "pink")

plt.show()
```

# Histograms

A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

Example: Say you ask for the height of 250 people, you might end up with a histogram like this:

```python
x = np.array([10, 20, 30, 40, 50, 60, 56, 45, 29])

plt.hist(x, ec='red')

plt.show()
```

# Pie Charts

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([20, 30, 50, 59, 78])

plt.pie(x)

plt.show()
```

As you can see the pie chart draws one piece (called a wedge) for each value in the array (in this case [35, 25, 25, 15]).

By default the plotting of the first wedge starts from the x-axis and moves counterclockwise:

Labels - 


Add labels to the pie chart with the label parameter.

The label parameter must be an array with one label for each wedge:

```python
# x = np.array([34, 45, 20, 15])

labels = ["Computer", "Civil", "Mechanical", "IT", "ETC"]

plt.pie(x, labels=labels)

plt.show()
```

Explode

Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:

```python

my_explode = [0, 0.04, 0, 0, 0.1]

plt.pie(x, labels=labels, explode=my_explode)

plt.show()
```

Shadow

Add a shadow to the pie chart by setting the shadows parameter to True:



```python

my_explode = [0, 0.04, 0, 0, 0.1]

plt.pie(x, labels=labels, explode=my_explode, shadow=True)

plt.show()
```

Colors


You can set the color of each wedge with the colors parameter.

The colors parameter, if specified, must be an array with one value for each wedge:

```python
# 'r' - Red
# 'g' - Green
# 'b' - Blue
# 'c' - Cyan
# 'm' - Magenta
# 'y' - Yellow
# 'k' - Black
# 'w' - White
```

```python

my_explode = [0, 0.04, 0, 0, 0.1]

my_colors = ['r', 'c', 'y', 'k', 'm']

plt.pie(x, labels=labels, explode=my_explode, shadow=True, colors=my_colors)

plt.show()
```

Legend

To add a list of explanation for each wedge, use the legend() function:



```python

my_explode = [0, 0.04, 0, 0, 0.1]

my_colors = ['r', 'c', 'y', 'w', 'm']

plt.pie(x, labels=labels, explode=my_explode, shadow=True, colors=my_colors)

plt.legend()

plt.show()
```

# Adding percentage to wedges

```python

my_explode = [0, 0.04, 0, 0, 0.1]

my_colors = ['r', 'c', 'y', 'w', 'm']

plt.pie(x, labels=labels, explode=my_explode, shadow=True, colors=my_colors, autopct='%1.2f%%')

plt.legend()

plt.show()
```
